```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('./../datasets/Barcelona/Barcelona_Datasets/births.csv')
```

```python
df.head()
```

```python
df = df[df['District Name']!='No consta']
```

```python
df1 = pd.pivot_table(data= df, values = 'Number', columns = ['Year', 'District Name'], aggfunc = np.sum)
```

```python
df1.T
```

```python
df1[2017].values[0].tolist()
```

```python
dists = list(df1[2017].columns)
```

```python
yrs = list(df['Year'].unique())
```

### Multi-line plot which tells us how the district birthrates change with respect to time

```python
fig,ax = plt.subplots(figsize = (20,8))
for i in yrs:
    ax.plot(dists, df1[i].values[0].tolist(), marker = 'o', linewidth = 2, label = str(i))
plt.xticks(rotation = 60)
plt.legend()
plt.title('District-wise birthrates change w.r.t Time in years', fontsize = 20)
plt.show()
```

### Inference:
- The birthrates of almost all the district have slowly increased with time.
- For a few districts, the birth rates have remained constant or have been uncertain with time.

### Scatterplot of NO2 vs O3 values in the Barcelona Air Quality Dataset of Nov 2017

```python
df3 = pd.read_csv('./../datasets/Barcelona/Barcelona_Datasets/air_quality_Nov2017.csv')
```

```python
df3
```

```python
df3['O3 Value'].isnull().sum()
```

```python
df3['NO2 Value'].isnull().sum()
```

```python
df3.dropna(subset = ['O3 Value', 'NO2 Value'], axis = 0, inplace = True)
```

```python
plt.figure(figsize = (8,5))
plt.scatter(df3['NO2 Value'],df3['O3 Value'],c = df3['NO2 Value'],cmap = 'jet', alpha = .5)
plt.xlabel('NO2 Value')
plt.ylabel('O3 Value')
plt.title('NO2 vs O3 relationship (ug/m3)', fontsize = 15)
plt.show()
```

### Inference
- There is a polynomial correlation between the NO2 and the O3 values
- The avg. concentration of O3 decreases when the concentration of NO2 increases
